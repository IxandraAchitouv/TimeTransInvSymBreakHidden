```python
# Dec. 2024 
# this notebooks can be used to make RG trajectories from 
#'Time_translation_invariance_symmetry_breaking_hidden_behind_finite_scale_singularities'
# and to make Fig4: Time_translation_invariance_symmetry_breaking_hidden_behind_finite_scale_singularities-1
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

from scipy.integrate import solve_ivp

def dim_2n(k, n):
    """Calculate dim_{2n}(k)."""
    return (n - 1) * (k**2 / (k**2 - 4)) + 2 * (2 - n)
    
def dim_6_nl(k):
    """Calculate dim_{6,n.l}(k) for the non-local coupling."""
    if k >= 2:  # Handle potential singularity for sin^{-1}(k/2)
        raise ValueError("k must be less than 2 to avoid singularity.")
    
    term1 = k * (k**2 - 8) * np.sqrt(4 - k**2)
    term2 = 4 * (8 - 3 * k**2) * np.arcsin(k / 2)
    denominator = (k**2 - 4) * (k * np.sqrt(4 - k**2) - 4 * np.arcsin(k / 2))
    
    return (term1 + term2) / denominator

def solve_system(bar_u2_0, bar_u4_0, bar_u6_0, bar_tilde_u6_0, t_span, t_eval):
    """
    Numerically solve the coupled ODEs for bar_u2, bar_u4, bar_u6, and bar_tilde_u6.
    
    Parameters:
    ----------
    bar_u2_0, bar_u4_0, bar_u6_0, bar_tilde_u6_0 : float
        Initial conditions for bar_u2, bar_u4, bar_u6, and bar_tilde_u6.
    t_span : tuple
        Time span as (start_time, end_time).
    t_eval : array
        Time points at which to evaluate the solution.
        
    Returns:
    -------
    t : array
        Time points.
    bar_u2, bar_u4, bar_u6, bar_tilde_u6 : arrays
        Solutions for bar_u2, bar_u4, bar_u6, and bar_tilde_u6 at the time points.
    """
    # Define constants
    pi = np.pi

    # Define the system of ODEs
    def odes(t, y):
        bar_u2, bar_u4, bar_u6, bar_tilde_u6 = y
        
        # Compute k as a function of t
        k = np.exp(-t)
        
        # Precompute dimensions
        dim_4 = dim_2n(k, n=2)
        dim_6 = dim_2n(k, n=3)
        dim_6_nl_value = dim_6_nl(k)
        
        # Define the derivatives
        dbar_u2_dt = -2 * bar_u2 - bar_u4 / (36 * pi)
        dbar_u4_dt = -(
            -dim_4 * bar_u4
            - bar_tilde_u6 / (30 * pi**2)
            - bar_u6 / (60 * pi)
            + bar_u4**2 / (12 * pi)
        )
        dbar_u6_dt = -(
            -dim_6 * bar_u6
            + (72 * bar_u4 * bar_u6) / (5 * pi)
            + (4 * bar_u4 * bar_tilde_u6) / (5 * pi**2)
            - (5 * bar_u4**3) / (18 * pi)
        )
        dbar_tilde_u6_dt = dim_6_nl_value * bar_tilde_u6
        
        return [dbar_u2_dt, dbar_u4_dt, dbar_u6_dt, dbar_tilde_u6_dt]
    
    # Initial conditions
    y0 = [bar_u2_0, bar_u4_0, bar_u6_0, bar_tilde_u6_0]
    
    # Solve the system
    sol = solve_ivp(odes, t_span, y0, t_eval=t_eval, method='RK45')
    
    return sol.t, sol.y[0], sol.y[1], sol.y[2], sol.y[3]  # Return bar_u2, bar_u4, bar_u6, and bar_tilde_u6




######## Fig 2:
def check_divergence(t, u4):
    """
    Check if u4 diverges numerically and return the value at a specified time point if not.
    
    Parameters:
    t : array
        Array of time points.
    u4 : array
        Array of u4 values.
    threshold : float
        The divergence threshold.
        
    Returns:
    u4_test : float or np.nan
        The value of u4 at the test time if no divergence, otherwise np.nan.
    """
    threshold=15
    time_limit_fraction=3./4
    # Set the time limit at 3/4 of the time range
    nt = int(time_limit_fraction * len(t))  # Use 3/4 of the total array
    time_limit = t[nt]
    
    # Mask for t > time_limit
    time_mask = t > time_limit
    
    # Check if any |u4| values after time_limit exceed the threshold
    u4_post_limit = u4[time_mask]
    if np.any(abs(u4_post_limit) > threshold):
        return np.nan  # Divergence detected, return NaN
    
    # No divergence detected, return the value of u4 at the test time
    u4_test = u4[nt]
    return u4_test


def check_divergencev2(t, u4, threshold=1e3, max_rate_change=1e6, increasing_rate_threshold=1e-2):
    """
    Check for divergence in the u4 array based on numerical stability criteria.
    
    Parameters:
    t : array-like
        Array of time points.
    u4 : array-like
        Array of u4 values.
    threshold : float, optional
        The threshold for identifying large values in the solution (default 1e3).
    max_rate_change : float, optional
        The maximum allowed rate of change (default 1e6), beyond which we consider divergence.
        
    Returns:
    u4_test : float or np.nan
        The value of u4 at the test time if no divergence, otherwise np.nan.
    """
    # Convert to numpy array if needed
    t = np.array(t)
    u4 = np.array(u4)
    
    # Check for NaN or infinite values in u4 (numerical instability)
    if np.any(np.isnan(u4)) or np.any(np.isinf(u4)):
        return np.nan  # Divergence detected
    
    # Check if any u4 values exceed the threshold (for potential instability)
    #if np.any(np.abs(u4) > threshold):
    #    return np.nan  # Divergence detected

    # Mask for t > time_limit
    nc=int(3*len(t)/4)
    time_limit=t[nc]
    time_mask = t > time_limit
    
    # Check if any |u4| values after time_limit exceed the threshold
    u4_post_limit = u4[time_mask]
    if np.any(abs(u4_post_limit) > threshold):
        return np.nan  # Divergence detected
    
    # Check the rate of change of u4 (if it changes too rapidly, it may indicate divergence)
    u4_diff = np.diff(u4) / np.diff(t)  # Rate of change of u4 over time
    if np.any(np.abs(u4_diff) > max_rate_change):
        return np.nan  # Divergence detected

    # Check the rate of change of u4 (u4_diff is the first derivative)
    u4_diff = np.diff(u4) / np.diff(t)  # First derivative (rate of change) of u4 over time
    
    # Check if the rate of change is constantly increasing
    rate_of_change_diff = np.diff(u4_diff)  # Second derivative (change in the rate of change)
    
    if np.all(rate_of_change_diff > increasing_rate_threshold):
        return np.nan  # Divergence detected due to constantly increasing rate of change
    
    # If no divergence detected, return the last valid value of u4
    u4_test = u4[-1]  # Or any other point you wish to return
    return u4_test

def mcmc_sampling(num_samples, t_span, t_eval):
    # becarefull here u_i is in fact ubar_i
    """Run MCMC sampling and keep the set of values (u4_0, u6_0, u6_tilde_0, u4_test)."""
    results = []  # List to store results for each sample
    
    # Sample initial conditions
    u4_samples, u6_samples, u6_tilde_samples = sample_initial_conditions(num_samples)
    u2_0=0.001
    for i in range(num_samples):
        # Get initial conditions for this sample
        u4_0, u6_0, u6_tilde_0 = u4_samples[i], u6_samples[i], u6_tilde_samples[i]
        
        try:
            # Solve the system
            t, u2,u4, u6, bar_tilde_u6 = solve_system(u2_0,u4_0, u6_0, u6_tilde_0, t_span, t_eval)
            
            # Check for divergence and get u4 at the test time
            u4_test = check_divergence(t, u4)
            
            # Store the result for this sample
            results.append((u4_0, u6_0, u6_tilde_0, u4_test))
        
        except Exception as e:
            print(f"Solver failed for initial conditions: u4_0={u4_0}, u6_0={u6_0}, u6_tilde_0={u6_tilde_0}. Error: {e}")

    return results


def sample_initial_conditions(num_samples, u6div=10000):
    """Generate random samples for initial conditions with constraints."""
    # Ensure u4_samples and u6_samples are always > 0
    #u4_samples = np.random.uniform(0.01, 1, num_samples)
    #u6_samples = np.random.uniform(0.1, 1.5, num_samples)
    

    # Split samples for u6_tilde_0
    #num_close_to_bounds = num_samples // 4  # 25% for clusters near -u6div and 0
    #num_uniform = num_samples - 2 * num_close_to_bounds  # Remaining 50% uniformly distributed

    # Uniform samples in [-u6div, 0]
    #u6_tilde_uniform = np.random.uniform(-u6div, 0, num_uniform)

    # Cluster around 0
    #u6_tilde_near_0 = np.random.normal(0, u6div / 4, num_close_to_bounds)
    #u6_tilde_near_0 = np.clip(u6_tilde_near_0, -u6div, 0)

    # Cluster around -u6div
    #u6_tilde_near_minus_u6div = np.random.normal(-u6div, u6div / 4, num_close_to_bounds)
    #u6_tilde_near_minus_u6div = np.clip(u6_tilde_near_minus_u6div, -u6div, 0)

    # Combine samples
    #u6_tilde_samples = np.concatenate([u6_tilde_uniform, u6_tilde_near_0, u6_tilde_near_minus_u6div])
    #u6_tilde_samples=np.random.uniform(-u6div, 0, num_samples)

    
    u4_samples = np.random.uniform(0.2, 0.4, num_samples)
    u6_samples = np.random.uniform(0.9, 1.1, num_samples)
    u6_tilde_samples=np.random.uniform(-u6div, 0, num_samples)
    
    return u4_samples, u6_samples, u6_tilde_samples



def plot_scatter(results):
    """
    Create three scatter plots for specific pairs of sampled parameters,
    coloring points by u4_test and using black for divergent samples.
    
    Parameters:
    results : list of tuples
        Output from `mcmc_sampling`, containing (u4_0, u6_0, u6_tilde_0, u4_test).
    """
    # Extract data from results
    u4_samples = np.array([res[0] for res in results])  # u4_0
    u6_samples = np.array([res[1] for res in results])  # u6_0
    u6_tilde_samples = np.array([res[2] for res in results])  # u6_tilde_0
    u4_test = np.array([res[3] for res in results])  # u4_test

    # Replace NaN values in u4_test with a placeholder for coloring (e.g., very low value)
    u4_test_color = np.copy(u4_test)
    diverged_mask = ~np.isfinite(u4_test_color)  # Mask for divergent cases
    u4_test_color[diverged_mask] = np.nanmin(u4_test) - 1  # Assign a distinct low value for divergence
    
    # Create figure
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    pairs = [
        (u6_samples, u4_samples, "$u6_0$", "$u4_0$"),
        (u4_samples, u6_tilde_samples, "$u4_0$", "$u6_{\tilde{0}}$"),
        (u6_samples, u6_tilde_samples, "$u6_0$", "$u6_{\tilde{0}}$"),
    ]

    # Loop through each pair of variables and create a scatter plot
    for ax, (x, y, xlabel, ylabel) in zip(axes, pairs):
        sc = ax.scatter(
            x, y, c=u4_test_color, cmap="viridis", s=20, alpha=0.7, edgecolor='k'
        )
        # Divergent points (NaN in original u4_test) are plotted in black
        ax.scatter(x[diverged_mask], y[diverged_mask], c="black", label="Diverged", s=20)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.legend(loc="upper right")
        ax.grid(True, linestyle="--", alpha=0.5)

    # Add a colorbar
    cbar = fig.colorbar(sc, ax=axes, orientation="vertical", shrink=0.8, pad=0.02)
    cbar.set_label("$u4_{\mathrm{test}}$", rotation=270, labelpad=15)
    cbar.set_ticks([np.nanmin(u4_test_color), np.nanmax(u4_test)])  # Use actual range of u4_test
    
    plt.tight_layout()
    plt.show()
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_scatter(results):
    """
    Create a 3D scatter plot for u4_samples, u6_samples, and u6_tilde_samples,
    coloring points by u4_test and using black for divergent samples.
    
    Parameters:
    results : list of tuples
        Output from `mcmc_sampling`, containing (u4_0, u6_0, u6_tilde_0, u4_test).
    """
    # Extract data from results
    u4_samples = np.array([res[0] for res in results])  # u4_0
    u6_samples = np.array([res[1] for res in results])  # u6_0
    u6_tilde_samples = np.array([res[2] for res in results])  # u6_tilde_0
    u4_test = np.array([res[3] for res in results])  # u4_test

    # Handle divergence: Divergent points (u4_test = NaN) will be black
    finite_mask = np.isfinite(u4_test)  # Mask for non-divergent points
    colors = np.zeros_like(u4_test, dtype=float)
    colors[finite_mask] = u4_test[finite_mask]  # Use u4_test for color
    colors[~finite_mask] = np.nanmin(u4_test) - 1  # Assign a distinct low value for divergence
    
    # Create figure
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Scatter plot
    scatter = ax.scatter(
        u4_samples, u6_samples, u6_tilde_samples, 
        c=colors, cmap="viridis", s=20, alpha=0.7, edgecolor='k'
    )
    # Highlight divergent points in black
    ax.scatter(
        u4_samples[~finite_mask], u6_samples[~finite_mask], u6_tilde_samples[~finite_mask], 
        c="black", label="Diverged", s=20
    )
    
    # Set labels$\\bar{{u}}_4(0)={u4_0}, \\bar{{u}}_6(0)={u6_0}, \\tilde{{\\bar{{u}}}}_6(0)
    ax.set_xlabel("$\\bar{{u}}_4(0)$")
    ax.set_ylabel("$\\bar{{u}}_6(0)$")
    ax.set_zlabel("$\\tilde{{\\bar{{u}}}}_6(0)$")
    
    # Add colorbar
    cbar = fig.colorbar(scatter, ax=ax, shrink=0.6, pad=0.1)
    cbar.set_label("$\\bar{{u}}_4(t\gg1)$", rotation=270, labelpad=15)
    cbar.set_ticks([np.nanmin(colors), np.nanmax(colors)])  # Use the range of u4_test
    
    plt.legend(loc="upper left")
    plt.savefig('fig3Ddiv.png')
    plt.show()






# Parameters
runexemple=0
if runexemple==1:
    u4bar_0 = 0.3  # Initial condition for u4
    u6bar_0 = 1.  # Initial condition for u6
    u6bar_tilde_0 = -50000  # Initial condition for bar_tilde_u6
    t_span = (-np.log(1.99), 3)  # Time from 0 to 10
    t_eval = np.linspace(-np.log(1.99), 3, 500)  # 500 points in time
    
    # Solve the system
    t, u4bar, u6bar, bar_tilde_u6 = solve_system(u4bar_0, u6bar_0, u6bar_tilde_0, t_span, t_eval)
    
        
        
        # Plotting the results (optional)
    plt.plot(t, u4bar, label='u4(t)')
    #plt.plot(t, u6bar, label='u6(t)')
    plt.xlabel('Time')
    plt.ylabel('Values')
    plt.legend()
    plt.title('Numerical Solution for u4 and u6')
    plt.grid(True)
    plt.show()

```
